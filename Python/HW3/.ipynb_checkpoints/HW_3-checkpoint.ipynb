Использовать можно что-угодно, но полученные данные должны быть +- актуальными на момент сдачи задания

````python
import requests 
import re
from bs4 import BeautifulSoup
````

````python
response = requests.get("https://www.imdb.com/chart/top/", params={"ref_": "nv_mp_mv250"})
soup = BeautifulSoup(response.content, "lxml")
````

Проще всего сначала выполнить последнюю часть, а потом все остальные:
Сохраните данные по всем 250 фильмам в виде таблицы с колонками (name, rank, year, rating, n_reviews, director) в любом формате (2 балла)

````python
import pandas as pd

titles = [soup.find_all("a", title=True)[i].text for i in range(250)]
rating_tags = soup.find_all(class_="ratingColumn imdbRating")
number_of_ratings = [int(rating_tags[i].find("strong")['title'].split()[-3].replace(',','')) for i in range(250)]
years_of_release = [int(soup.find_all("span", class_="secondaryInfo")[i].text.replace('(','').replace(')', ''))\
                    for i in range(250)]
ratings = [float(rating_tags[i].text.replace('\n', '')) for i in range(250)]
directors = [soup.find_all("a", title=True)[i]["title"].split('(dir.)')[0][:-1] for i in range(250)]

top_250_df = pd.DataFrame({"title": titles, "number_of_ratings": number_of_ratings,\
                           "year_of_release": years_of_release, "rating": ratings, "director": directors})

top_250_df.to_csv('imbd_top_250.csv', index=True)
````

````python
top_250_df.head(10)
````

Выведите топ-4 фильма по количеству оценок пользователей и количество этих оценок 

````python
top_250_df.sort_values("number_of_ratings")[::-1][:4]
````

````python
grouped_by_year = top_250_df.groupby("year_of_release")
````

Выведите топ-4 лучших года (по среднему рейтингу фильмов в этом году) и средний рейтинг (1 балл)

````python
grouped_by_year.mean()['rating'].sort_values()[::-1][:4]
````

Постройте отсортированный barplot, где показано количество фильмов из списка для каждого режисёра (только для режиссёров с более чем 2 фильмами в списке) (1 балл)

````python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set()

directors = top_250_df.groupby('director').filter(lambda x: len(x) > 1).groupby('director').size().sort_values()
directors.plot(kind='barh')

plt.xlabel('Number of Films')
plt.ylabel('Director')
plt.tick_params(axis='y', which='major', labelsize=7)
plt.title('Number of Films per Director')

plt.show()
````

Выведите топ-4 самых популярных режиссёра (по общему числу людей оценивших их фильмы) (2 балла)

````python
top_250_df.groupby("director").sum("number_of_ratings").sort_values("number_of_ratings")[::-1][:4]
````

Результат запуска кода ниже должен быть примерно такой:







````python
import sys
import time
import os
import io
import requests
import dotenv
from dotenv import dotenv_values

TOKEN = dotenv_values(".env")['TG_API_TOKEN']
API_URL = "https://api.telegram.org/bot" + TOKEN + "/"

#    Декоратор должен принимать один обязательный аргумент — ваш CHAT_ID в телеграме. Как узнать свой CHAT_ID можно найти в интернете
#    В сообщении об успешно завершённой функции должны быть указаны её имя и время выполнения
#    В сообщении о функции, завершившейся с исключением, должно быть указано имя функции, тип и текст ошибки
#    Ключевые элементы сообщения должны быть выделены как код (см. скриншот), форматирование остальных элементов по вашему желанию
#    Время выполнения менее 1 дня отображается как HH:MM:SS.μμμμμμ, время выполнения более 1 дня как DDD days, HH:MM:SS. Писать форматирование самим не нужно, всё уже где-то сделано за вас

#  К сообщению также должен быть прикреплён файл, содержащий всё, что декорируемая функция записывала в stdout и stderr во время выполнения. Имя файла это имя декорируемой функции с расширением .log (+3 дополнительных балла)
#  Реализовать предыдущий пункт, не создавая файлов на диске (+2 дополнительных балла)
#  Если функция ничего не печатает в stdout и stderr — отправлять файл не нужно

CHAT_ID = -998612919

import requests
from datetime import datetime
import time
import sys
import io

def post(message, chat_id):
    requests.post(API_URL + "sendMessage", params={"chat_id": chat_id, "text": message, "parse_mode": "MarkdownV2"})

def send_log(chat_id, text, log_name):
    requests.post(API_URL + "sendDocument", data={"chat_id": chat_id}, files={"document": (f"{log_name}.txt", text, "text/plain")})

def telegram_logger(chat_id):
    def decorator(func):
        def inner_func(*args, **kwargs):
            func_name = func.__name__
            
            launch_datetime = datetime.now()
            lt_string = launch_datetime.strftime("%d/%m/%Y %H:%M:%S")

            # Redirect stdout and stderr to a StringIO buffer
            original_stdout = sys.stdout
            original_stderr = sys.stderr
            sys.stdout = io.StringIO()
            sys.stderr = io.StringIO()

            try:
                result = func(*args, **kwargs)
                success = True
            except Exception as e:
                success = False
                error_message = str(e)

            # Restore stdout and stderr, and get the buffer content
            stdout_content = sys.stdout.getvalue()
            stderr_content = sys.stderr.getvalue()
            sys.stdout = original_stdout
            sys.stderr = original_stderr
            
            # Send the log if it is not empty
            if stdout_content or stderr_content:
                log_content = f"stdout:\n{stdout_content}\nstderr:\n{stderr_content}"
                send_log(chat_id, log_content, func_name)

            finish_datetime = datetime.now()
            ft_string = finish_datetime.strftime("%d/%m/%Y %H:%M:%S")
            elapsed_time = finish_datetime - launch_datetime
            total_seconds = elapsed_time.total_seconds()
            
            # A bit of ChatGPT magic
            if elapsed_time.days == 0:
                ex_string = f"{int(total_seconds // 3600):02d}:{int((total_seconds // 60) % 60):02d}:{total_seconds % 60:06.3f}"
            else:
                ex_string = f"{elapsed_time.days} days {elapsed_time.seconds // 3600:02d}:{(elapsed_time.seconds // 60) % 60:02d}:{elapsed_time.seconds % 60:02d}"

            if success:
                post(f'Function `{func_name}` launched at `{lt_string}`, successfully finished at `{ft_string}`\n'
                     f'Execution time compromised `{ex_string}`', chat_id)
            else:
                post(f"Function `{func_name}` launched at `{lt_string}` exited with the following error message: \n ``` \n {error_message} \n ```", chat_id)

            return result

        return inner_func
    return decorator


@telegram_logger(CHAT_ID)
def good_function():
    print("This goes to stdout")
    print("And this goes to stderr", file=sys.stderr)
    time.sleep(2)
    print("Wake up, Neo")

@telegram_logger(CHAT_ID)
def bad_function():
    print("Some text to stdout")
    time.sleep(2)
    print("Some text to stderr", file=sys.stderr)
    raise RuntimeError("Ooops, exception here!")
    print("This text follows exception and should not appear in logs")
    
@telegram_logger(CHAT_ID)
def long_lasting_function():
    time.sleep(200000000)


good_function()

try:
    bad_function()
except Exception:
    pass

# long_lasting_function()
````

````python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from typing import IO, Union, Literal
````

````python
@dataclass
class GenscanOutput:
    status: str
    cds_list: str
    intron_list: pd.DataFrame
    exon_list: pd.DataFrame

def run_genscan(sequence: str = None,
                sequence_file: Union[str, IO] = None,
                organism: Literal["Vertebrate", "Arabidopsis", "Maze"] = "Vertebrate",
                exon_cutoff: float = 1.00,
                sequence_name: str = "") -> GenscanOutput:
                
        FORM_URL = "http://hollywood.mit.edu/cgi-bin/genscanw_py.cgi"
        
        # Choose if the form is to be filled using a file or "manual" input
        if sequence_file:
            data = {"-o":organism,
                    "-e":exon_cutoff,
                    "-n":sequence_name,
                    "-p":"Predicted peptides only",
                    "-u":sequence_file}
        else:
            data = {"-o":organism,
                    "-e":exon_cutoff,
                    "-n":sequence_name,
                    "-p":"Predicted peptides only",
                    "-s":sequence}
            
        
        response = requests.post(url=FORM_URL, data=data)
        status = response.status_code
        
        soup = BeautifulSoup(response.content, "lxml")
        text = soup.find_all("pre").__repr__()
          
        # A tiny itsy-bitsy little humble bit of GPT-4 diabolic sorcery below
        
        # Extract exons and introns
        exon_intron_pattern = r"(\d+\.\d+)\s+(Intr|Term|PlyA)\s+(\+|\-)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+)?\s+(\S+)?\s+(\S+)?"
        matches = re.findall(exon_intron_pattern, text)

        exon_data = []
        intron_data = []

        for match in matches:
            match = tuple(x if x is not None else '' for x in match)
            if match[1] == "Intr":
                intron_data.append(match)
            elif match[1] in ["Term", "PlyA"]:
                exon_data.append(match[:-1] + ('',))

        # Make sure each row has the correct length
        exon_data = [row if len(row) == 12 else row + ('',) for row in exon_data]
        intron_data = [row if len(row) == 13 else row + ('',) for row in intron_data]

        columns = ["ID", "Type", "Strand", "Begin", "End", "Length", "Frame", "Phase", "I/Ac", "Do/T", "CodRg", "P", "Tscr"]
        exon_df = pd.DataFrame(exon_data, columns=columns[:-1])
        intron_df = pd.DataFrame(intron_data, columns=columns)

        # Extract amino acid sequence
        amino_acid_pattern = r"(?<=predicted_peptide_1\|)(\d+)_aa\n\n(.*\n)*"
        amino_acid_match = re.search(amino_acid_pattern, text, flags=re.MULTILINE)
        amino_acid_sequence = amino_acid_match.group(0).replace("\n", "") if amino_acid_match else ""

        exon_list = exon_df
        intron_list = intron_df
        cds_list = amino_acid_sequence.split("_aa")[1]

        return GenscanOutput(status=status, cds_list=cds_list, intron_list=intron_list, exon_list=exon_list)
        
````
