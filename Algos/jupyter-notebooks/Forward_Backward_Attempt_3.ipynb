```python
import numpy as np

inp = input().split()

pi = np.array(list(map(float, inp[:2])))

O = np.array(list(map(int, [*inp[10]])))

A = np.array([[float(inp[2]), float(inp[3])],\
              [float(inp[4]), float(inp[5])]])

B = np.array([[float(inp[6]),float(inp[7])],\
              [float(inp[8]),float(inp[9])]])

print(pi)
print(O)
print(A)
print(B)
print('=====================\n \n=====================')

def forward_backward_2(A, B, pi, O):

    T = len(O)
    N = 2
    
    a = np.zeros((2, T))
    a[:, 0] = [pi[j]*B[j,O[0]] for j in range(N)]
    for t in range(1, T):
        for j in range(2):
            a[j, t] = sum(a[i, t-1]*A[i, j]*B[j, O[t]] for i in range(N))
            
    print(a.T)
    print('++++++++++++++++')

    b = np.zeros((2, T))
    b[:, T-1] = [1, 1]
    for t in range(T-1)[::-1]:
        for i in range(2):
            b[i, t] = sum(A[i, j]*B[j,O[t+1]]*b[j, t+1] for i in range(N)) 
            
    print(b.T)
    print('++++++++++++++++')

    P = np.zeros((2, T))
    for t in range(T):
        for j in range(2):
            P[j, t] = (a[j, t]*b[j, t])/sum([a[i, T-1] for i in range(N)])
            
    return P.T


print(forward_backward_2(A, B, pi, O))

```
